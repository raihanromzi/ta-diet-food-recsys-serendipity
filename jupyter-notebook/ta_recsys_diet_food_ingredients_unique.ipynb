```python
import pandas as pd
import json
```

```python
df = pd.read_csv('./dataset/preprocessed_recipes.csv')
df.head(2)
```

```python
df_ingredients = df['RecipeIngredientPartsClean']
df_ingredients
```

```python
# Function to extract and clean unique ingredients
def extract_unique_ingredients(ingredients_series):
    # Split ingredients into lists, remove extra spaces, and convert to lowercase
    ingredients_list = ingredients_series.str.split(',').apply(lambda x: [i.strip().lower() for i in x])

    # Flatten the list of lists and get unique ingredients
    unique_ingredients = set(ingredient for sublist in ingredients_list for ingredient in sublist)

    return unique_ingredients
```

```python
# Extract unique ingredients
unique_ingredients = extract_unique_ingredients(df_ingredients)
```

```python
# Convert the set of unique ingredients to a sorted list
unique_ingredients_list = sorted(list(unique_ingredients))
```

```python
# If you want to display or save the unique ingredients for the frontend, you can convert them to a JSON or any other required format
unique_ingredients_json = json.dumps(unique_ingredients_list, indent=2)
print(unique_ingredients_json)
```

```python
with open('unique_ingredients.json', 'w') as f:
  f.write(unique_ingredients_json)
```
